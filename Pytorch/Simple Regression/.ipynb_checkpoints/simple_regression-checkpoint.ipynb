```python
# Import dependencies
import numpy as np
import torch
import matplotlib.pyplot as plt
```

```python
print(torch.__version__)
```

```python
# Generate some data
x = torch.unsqueeze(torch.linspace(-1, 1, 100), dim=1)
print(x)
print(x.shape)
```

```python
y = x.pow(2) + 0.2 * torch.rand(x.size())
print(y)
print(y.shape)
```

```python
# Plot the data
plt.scatter(x.data.numpy(), y.data.numpy())
plt.show()
```
